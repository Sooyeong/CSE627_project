```python
def rmse_cv_train(model):
    rmse=np.sqrt(-cross_val_score(model, X_train, y_train, scoring=scorer, cv=5))
    return (rmse)

def rmse_cv_test(model):
    rmse=np.sqrt(-cross_val_score(model, X_test, y_test, scoring=scorer, cv=5))
    return(rmse)
```
